
# Decision Tree Classifier

The classification is performed for iris dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets, tree   # The tree module contains Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()

X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state = 87652 )

DTree = DecisionTreeClassifier(max_depth=4)
DTree.fit(X_train, y_train)
```

## Visualizing Decision Tree

To visualize the fitted decision tree, we need to use a visualization tool that supports creation of decision trees. 

### Graphviz

Graphviz is open source graph visualization software. It can generate a visualization of graph described in a specially designed DOT language.

The tree module of Scikit-Learn provides a function `export_graphviz` that export a decision tree in DOT format.

#### Prerequisites
1. Install the latest version of Graphviz software from https://www.graphviz.org/download/.
2. Add the bin folder of Graphviz installation in PATH variable.
3. Install `graphviz` package using `pip install graphviz` command.

```python
import graphviz 
dot_data = tree.export_graphviz(DTree, out_file=None, 
                         feature_names=iris.feature_names,  
                         class_names=iris.target_names,  
                         filled=True, rounded=True,  
                         special_characters=True) 
graph = graphviz.Source(dot_data) 
graph
```

```python
y_pred = DTree.predict(X_test)
print('Accuracy = %0.4f'%np.mean(y_pred == y_test))
```
